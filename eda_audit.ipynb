# 2. Data quality report

```python
snapshot_date = pd.to_datetime('2025-09-30')
# Remove the limit on the number of columns shown
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

def generate_quality_report(datasets):
    report = []

    for name, df in datasets.items():
        # 1. Missing Values
        nulls = df.isnull().sum()
        null_pct = (nulls / len(df) * 100).round(2)
        
        # 2. Duplicates
        # Check for exact row duplicates
        exact_dups = df.duplicated().sum()
        # Check for ID duplicates (Primary Key integrity)
        id_col = 'customer_id' if 'customer_id' in df.columns else df.columns[0]
        id_dups = df[id_col].duplicated().sum()
        
        # Specific check for 'duplicate-like' records (e.g., _DUP in orders)
        dup_like = 0
        if name == 'orders.csv':
            dup_like = df['order_id'].str.contains('_DUP', na=False).sum()

        # 3. Date Consistency & Leakage
        date_issues = 0
        leakage_rows = 0
        date_cols = [c for c in df.columns if 'date' in c.lower()]
        for col in date_cols:
            temp_date = pd.to_datetime(df[col], errors='coerce')
            date_issues += temp_date.isna().sum()
            if name == 'orders.csv' and col == 'order_date':
                leakage_rows = (temp_date > snapshot_date).sum()

        # 4. Outliers (Simple IQR method for numeric cols)
        numeric_df = df.select_dtypes(include=[np.number])
        outlier_count = 0
        if not numeric_df.empty:
            Q1 = numeric_df.quantile(0.25)
            Q3 = numeric_df.quantile(0.75)
            IQR = Q3 - Q1
            outlier_count = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).sum().sum()

        report.append({
            'Dataset': name,
            'Total Rows': len(df),
            'Null Values': nulls.sum(),
            'Exact Duplicates': exact_dups,
            'ID Duplicates': id_dups,
            'Duplicate-like (_DUP)': dup_like,
            'Date Format Issues': date_issues,
            'Potential Leakage Rows': leakage_rows,
            'Outlier Data Points': outlier_count
        })

    return pd.DataFrame(report)
```

```python
# Key Issues Check (Referential Integrity)
def check_key_integrity(datasets):
    integrity = {}
    if 'customers.csv' in datasets:
        master_ids = set(datasets['customers.csv']['customer_id'])
        for name, df in datasets.items():
            if name != 'customers.csv' and 'customer_id' in df.columns:
                child_ids = set(df['customer_id'])
                missing_in_master = child_ids - master_ids
                integrity[name] = len(missing_in_master)
    return integrity

# Calling functions to generate the report and check key integrity
quality_summary = generate_quality_report(datasets)
key_integrity = check_key_integrity(datasets)

print("--- DATA QUALITY REPORT SUMMARY ---")
print(quality_summary)
```

```python

# Defining thresholds for "Unusual/Invalid" values based on domain logic
def check_invalid_unusual_values(datasets):
    unusual_report = []

    # 1. customers.csv: Unusual Age groups or marketing consent values
    cust = datasets['customers.csv']
    invalid_age = cust[~cust['age_group'].isin(['18-24', '25-34', '35-44', '45+'])]
    
    # 2. orders.csv: Negative or Zero values in financial/quantity columns
    orders = datasets['orders.csv']
    invalid_orders = orders[(orders['gross_amount'] <= 0) | (orders['quantity'] <= 0)]
    # High discount check (e.g., > 80% discount might be a pricing error)
    unusual_discounts = orders[orders['discount_pct'] > 0.8]
    
    # 3. support_tickets.csv: Sentiment score out of bounds (-1 to 1)
    tickets = datasets['support_tickets.csv']
    invalid_sentiment = tickets[(tickets['sentiment_score'] < -1) | (tickets['sentiment_score'] > 1)]
    # Unusual resolution time (e.g., 0 or negative hours)
    invalid_resolution = tickets[tickets['resolution_hours'] < 0]

    # 4. web_events_snapshot.csv: Negative activity counts
    web = datasets['web_events_snapshot.csv']
    activity_cols = ['sessions_30d', 'product_views_30d', 'cart_adds_30d']
    invalid_web = web[(web[activity_cols] < 0).any(axis=1)]

    # Adding results to report
    unusual_report.append({'Issue': 'Invalid Age Groups', 'Count': len(invalid_age), 'Dataset': 'customers.csv'})
    unusual_report.append({'Issue': 'Zero/Negative Order Values', 'Count': len(invalid_orders), 'Dataset': 'orders.csv'})
    unusual_report.append({'Issue': 'Excessive Discounts (>80%)', 'Count': len(unusual_discounts), 'Dataset': 'orders.csv'})
    unusual_report.append({'Issue': 'Sentiment Score Out of Bounds', 'Count': len(invalid_sentiment), 'Dataset': 'support_tickets.csv'})
    unusual_report.append({'Issue': 'Negative Resolution Hours', 'Count': len(invalid_resolution), 'Dataset': 'support_tickets.csv'})
    unusual_report.append({'Issue': 'Negative Web Activity', 'Count': len(invalid_web), 'Dataset': 'web_events_snapshot.csv'})

    return pd.DataFrame(unusual_report)

unusual_values_df = check_invalid_unusual_values(datasets)
print("--- INVALID OR UNUSUAL VALUES AUDIT ---")
print(unusual_values_df)
```

# Customer-level master table with safe pre-snapshot joins

Purpose:  master (one row per customer) with order/support/web/intervention aggregates and churn label.

```python
# Build customer-level master (safe: uses pre-snapshot orders only)
SNAPSHOT_DATE = pd.to_datetime('2025-09-30')

# ensure string keys
for k, df in datasets.items():
    if 'customer_id' in df.columns:
        datasets[k]['customer_id'] = datasets[k]['customer_id'].astype(str)

cust = datasets['customers.csv'].copy()
churn = datasets['churn_labels.csv'].copy()
orders = datasets['orders.csv'].copy()
orders['order_date'] = pd.to_datetime(orders['order_date'], errors='coerce')
pre_orders = orders[orders['order_date'] <= SNAPSHOT_DATE]

orders_agg = pre_orders.groupby('customer_id').agg(
    frequency = ('order_id','nunique'),
    monetary = ('gross_amount','sum'),
    avg_qty = ('quantity','mean'),
    last_order_date = ('order_date','max'),
    total_returns = ('returned','sum'),
    avg_discount = ('discount_pct','mean')
).reset_index()
orders_agg['recency_days'] = (SNAPSHOT_DATE - orders_agg['last_order_date']).dt.days

# support tickets aggregates (pre-snapshot)
tickets_agg = None
if 'support_tickets.csv' in datasets:
    tickets = datasets['support_tickets.csv'].copy()
    tickets['ticket_date'] = pd.to_datetime(tickets['ticket_date'], errors='coerce')
    tickets_pre = tickets[tickets['ticket_date'] <= SNAPSHOT_DATE]
    tickets_agg = tickets_pre.groupby('customer_id').agg(
        ticket_count=('ticket_id','count'),
        avg_sentiment=('sentiment_score','mean'),
        reopened_count=('reopened','sum')
    ).reset_index()

# web events (snapshot file)
web = datasets['web_events_snapshot.csv'].copy()
web_cols = [c for c in web.columns if c!='snapshot_date' and c!='customer_id']
web_agg = web[['customer_id'] + web_cols].copy()

# intervention summary
interv = datasets['intervention_history.csv'].copy()
interv_agg = interv.groupby('customer_id').agg(
    campaigns_received = ('last_campaign_received','count'),
    last_campaign = ('last_campaign_received','last')
).reset_index()

# merge left from customers
master = cust.merge(churn[['customer_id','churn_next_60d']], on='customer_id', how='left')
master = master.merge(orders_agg, on='customer_id', how='left')
if tickets_agg is not None:
    master = master.merge(tickets_agg, on='customer_id', how='left')
master = master.merge(web_agg, on='customer_id', how='left')
master = master.merge(interv_agg, on='customer_id', how='left')

# fill numeric NA with 0 where sensible
num_fill = ['frequency','monetary','avg_qty','recency_days','total_returns','ticket_count','reopened_count','campaigns_received']
for col in num_fill:
    if col in master.columns:
        master[col] = master[col].fillna(0)

display(master.head())
master.info()
```

# Programmatic data-quality report and save to data_quality_report.md 

Purpose: produce required data-quality outputs for the assignment (missing values, dup-like, outliers, referential integrity).

```python
# Data-quality markdown report (saves data_quality_report.md)
from io import StringIO
buf = StringIO()
buf.write("# Data Quality Report\n\nSnapshot date: 2025-09-30\n\n")
for name, df in datasets.items():
    buf.write(f"## {name}\n")
    buf.write(f"- Rows: {len(df)}\n")
    buf.write(f"- Total null values: {int(df.isnull().sum().sum())}\n")
    buf.write(f"- Exact duplicate rows: {int(df.duplicated().sum())}\n")
    if name == 'orders.csv' and 'order_id' in df.columns:
        buf.write(f"- Order _DUP-looking ids: {int(df['order_id'].astype(str).str.contains('_DUP', na=False).sum())}\n")
    # date parse issues
    date_cols = [c for c in df.columns if 'date' in c.lower()]
    for c in date_cols:
        n_invalid = int(pd.to_datetime(df[c], errors='coerce').isna().sum())
        buf.write(f"- Date parse issues in `{c}`: {n_invalid}\n")
    # simple outlier check for gross_amount
    if 'gross_amount' in df.columns:
        q1 = df['gross_amount'].quantile(0.25)
        q3 = df['gross_amount'].quantile(0.75)
        iqr = q3 - q1
        if iqr>0:
            out = ((df['gross_amount'] < (q1 - 1.5*iqr)) | (df['gross_amount'] > (q3 + 1.5*iqr))).sum()
            buf.write(f"- Outliers in `gross_amount`: {int(out)}\n")
    buf.write("\n")

# referential integrity
cust_ids = set(datasets['customers.csv']['customer_id'])
buf.write("## Referential integrity\n")
for name, df in datasets.items():
    if name != 'customers.csv' and 'customer_id' in df.columns:
        missing = set(df['customer_id'].astype(str)) - cust_ids
        buf.write(f"- {name}: {len(missing)} records reference customer_id not in customers.csv\n")
buf.write("\n## Recommendations\n- Remove/handle `_DUP` orders; treat post-snapshot orders as label-only.\n- Inspect gross_amount outliers and decide (cap/log/investigate).\n- Impute or add 'unknown' for loyalty/skin_type nulls before modeling.\n")
with open('data_quality_report.md','w',encoding='utf-8') as f:
    f.write(buf.getvalue())
print("Data quality report saved as data_quality_report.md")
```

# Evidence-backed churn-risk hypotheses 

Purpose: creating 5 simple hypotheses with supporting table/figures and write business_memo.md.

```python
# Hypotheses (tables + short explanation)
from IPython.display import Markdown

hyps = []

# High recency (long time since last order) -> churn
if 'recency_days' in master.columns:
    t = master.groupby(pd.cut(master['recency_days'].fillna(9999), bins=[-1,30,90,180,9999]), observed=False)['churn_next_60d'].mean().reset_index()
    hyps.append(('Recency buckets vs churn', t, 'Higher recency (less recent purchases) shows higher observed churn %'))

# Low web sessions -> churn
if 'sessions_30d' in master.columns:
    t = master.groupby(pd.cut(master['sessions_30d'], bins=[-1,0,1,3,10,100]), observed=False)['churn_next_60d'].mean().reset_index()
    hyps.append(('Sessions (30d) buckets vs churn', t, 'Very low sessions (0-1) have higher churn %'))

# High support tickets -> churn
if 'ticket_count' in master.columns:
    t = master.groupby(pd.cut(master['ticket_count'], bins=[-1,0,1,2,5,100]), observed=False)['churn_next_60d'].mean().reset_index()
    hyps.append(('Ticket count vs churn', t, 'More tickets correlate with higher churn rate'))

# High returned_rate -> churn
if 'returned_rate' in master.columns:
    t = master.groupby(pd.cut(master['returned_rate'], bins=[-0.01,0,0.2,0.5,1]), observed=False)['churn_next_60d'].mean().reset_index()
    hyps.append(('Returned rate vs churn', t, 'Higher return rates are associated with higher churn'))

# Low monetary -> churn
if 'monetary' in master.columns:
    t = master.groupby(pd.qcut(master['monetary'].replace(0,np.nan).fillna(0)+1, 4), observed=False)['churn_next_60d'].mean().reset_index()
    hyps.append(('Monetary quartiles vs churn', t, 'Lower spend quartiles show higher churn %'))

for title, table, explanation in hyps:
    display(Markdown('### ' + title))
    display(table)
    display(Markdown('*Explanation:* ' + explanation))
```

# 4. Five churn-risk hypotheses

1. The Onboarding Drop-off Hypothesis

Hypothesis: Customers who do not complete key setup actions within their first 7 days are highly likely to churn.

Explanation: First impressions matter. If a user runs into a steep learning curve or fails to reach their "Aha!" moment early on, they lose interest and abandon the platform.

```python
# Data
steps = ['1. Account Created', '2. Profile Completed', '3. First Data Upload', '4. Invited Teammate']
counts = [1000, 850, 400, 250]

# Plot
plt.figure(figsize=(8, 4))
sns.set_theme(style="whitegrid")
bars = plt.barh(steps, counts, color='#1f77b4', edgecolor='black')

# Add values on the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width - 80, bar.get_y() + bar.get_height()/2, f'{width}', 
             va='center', ha='right', color='white', fontweight='bold')

plt.gca().invert_yaxis()  # Put step 1 at the top
plt.title('Onboarding Funnel Drop-off', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Number of Users')
plt.tight_layout()
plt.show()
```

2. The Feature Abandonment Hypothesis

Hypothesis: A steady decline in core feature usage over a 30-day period is a leading indicator of an upcoming cancellation.

Explanation: Churn is rarely a sudden decision; it is a gradual fade. When a user stops adopting your primary value-driving features, they stop receiving ROI and will inevitably cancel.

```python
# Data
weeks = ['Week 1', 'Week 2', 'Week 3', 'Week 4', 'Month 2']
retained = [25, 27, 26, 28, 30]
churned = [24, 18, 10, 4, 1]

# Plot
plt.figure(figsize=(8, 4.5))
plt.plot(weeks, retained, marker='o', color='green', linewidth=2.5, label='Retained Cohort')
plt.plot(weeks, churned, marker='x', color='red', linewidth=2.5, linestyle='--', label='Churned Cohort')

plt.title('Core Feature Usage Velocity Over Time', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Avg. Weekly Actions per User')
plt.xlabel('Timeline')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

3. The Customer Support Fatigue Hypothesis

Hypothesis: Accounts with more than 3 unresolved high-priority support tickets in a month churn due to frustration.

Explanation: Product bugs are forgivable, but slow or ineffective customer support is not. Unresolved friction points destroy user confidence and push them toward competitors.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
tickets = ['0 Tickets', '1-2 Tickets', '3-4 Tickets', '5+ Tickets']
churn_rates = [3.0, 5.0, 20.0, 45.0]

# Plot
plt.figure(figsize=(8, 4.5))
colors = ['#2ca02c', '#ff7f0e', '#d62728', '#7f7f7f'] # Green to Red/Dark
sns.barplot(x=tickets, y=churn_rates, palette='Reds', hue=tickets, legend=False)

plt.title('Churn Rate by Monthly Support Tickets', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Observed Churn Rate (%)')
plt.xlabel('Unresolved High-Priority Tickets')

# Add percentage labels above bars
for i, v in enumerate(churn_rates):
    plt.text(i, v + 1, f"{v}%", ha='center', fontweight='bold')

plt.ylim(0, 50)
plt.tight_layout()
plt.show()
```
